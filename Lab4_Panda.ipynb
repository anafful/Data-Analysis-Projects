```python
# Returns information about the dataframe
# Column names, number of non-null values and data types
df.info()

# Returns the first five rows of the dataframe
df.head()

# Returns the n-first rows of the dataframe
df.head(10)

# Returns the number of rows and columns as a tuple
df.shape

# Returns statistical analysis on appropriate columns
df.describe()

# Returns the mean of a dataframe -> not working
# because all columns are not numerical
df['mark'].mean()
```

Selecting Data

Columns

```python
# Return only one column
df['class']
print(type(df['class']))

# Return more than one column
# We need to use double square brackets because
# return value is dataframe
df[['class', 'gender']]

# Selecting rows with iloc
df.iloc[2] # one row at index 2
df.iloc[2:4] # rows starting at index 2, up until index 4 (excluded)

# Select columns with iloc
df.iloc[:, 0] # all rows and one column at index 0
df.iloc[:, 0:2] # all rows and two first columns

# Select rows and columns with iloc
df.iloc[0:2, 0:2] # first 2 rows and first two columns
df.iloc[::2, 0:2] # every other rows, first two columns
```

```python
# Selecting rows with loc
df.loc[2]
df.loc[2:4]

# Select columns with loc
df.loc[:, 'mark'] # one column -> no double square brackets
df.loc[:, 'id':'class'] # multiple columms, slice notation -> stop is included
df.loc[:, ['name', 'mark']] # multiple columns, list notation -> double square brackets
```

```python
# Check for null values in a Series
df['class'].isnull()

# Find the null values
df[df['class'].isnull()]

# Find the number of null values in a column
df['class'].isnull().sum()

# Find the number of null values per column in a dataframe
df.isnull().sum()

# Find the total number of null values in a dataframe
df.isnull().sum().sum()
```

```python
df['passed'] = df['mark'] > 60

##You dont need to create variable to define its duplicate of the table to make changes
df_score = df.rename(columns={'mark': 'score'})

df_score.head()
df_score.drop(columns=['passed'])
df_score.head()
```

```python
df_score['class'].value_counts() # selecting column [] and function
df_score.groupby('gender')['score'].mean() #selecting the function and column and selecting another column to find average
```
